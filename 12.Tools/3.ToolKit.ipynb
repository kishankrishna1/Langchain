## ToolKit

```python
from langchain_core.tools import tool

@tool
def add(a: int, b: int) -> int:
    """ Add two number """
    return a+b

@tool
def multiply(a:int, b:int) -> int:
    """ Multiply two number """
    return a*b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add, multiply]
```

```python
toolkit = MathToolKit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
    print(tool.name, " => ", tool.description)
```
