```python
from langchain_core.documents import Document
from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams
from langchain_openai import OpenAIEmbeddings
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
embeddings  = OpenAIEmbeddings(model="text-embedding-3-small"),
```

```python
# ── Create a LOCAL persistent Qdrant client ──────────────
client = QdrantClient(path="./qdrant_ipl_db")
```

```python
COLLECTION_NAME = "ipl_players"
```

```python
# ── Create the collection (only once) ─────────────────────────────────────────
client.recreate_collection(
    collection_name=COLLECTION_NAME,
    vectors_config=VectorParams(
        size=384,           # must match your embedding model's output size
        distance=Distance.COSINE
    )
)
```

```python
# ── Your IPL documents ─────────────────────────────────────────────────────────
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history...",
    metadata={"team": "Royal Challengers Bangalore", "doc_id": "doc1"}
)
doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history...",
    metadata={"team": "Mumbai Indians", "doc_id": "doc2"}
)
doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings...",
    metadata={"team": "Chennai Super Kings", "doc_id": "doc3"}
)
doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket...",
    metadata={"team": "Mumbai Indians", "doc_id": "doc4"}
)
doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball...",
    metadata={"team": "Chennai Super Kings", "doc_id": "doc5"}
)

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
# ── ADD DOCUMENTS (embeds + stores in one call) ────────────────────────────────
vectorstore = QdrantVectorStore.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name=COLLECTION_NAME,
    path="./qdrant_ipl_db",
)

# print(f"✅ {len(docs)})
```
